```python
o = np.arange(1, 16, 2)
e = np.arange(0, 16, 2)
fig, ax = plt.subplots()
ax.plot(o, np.zeros_like(o) + 1 , 'o')
ax.plot(e, 2 / (e + 2 ) , 'o')
ax.grid()
plt.show()


o1 = np.arange(1, 1600, 2)
e1 = np.arange(0, 1600, 2)
fig, ax1 = plt.subplots()
ax1.plot(o1, np.zeros_like(o1) + 1 , 'o')
ax1.plot(e1, 2 / (e1 + 2 ) , 'o')
ax1.grid()
plt.show()
```

```python
# Data for plotting
t = np.arange(0.1, 6.0, .001)
s =  t * np.sin(np.pi / t)
fig, ax = plt.subplots()
ax.plot(t, s)
ax.grid()


t1 = np.arange(0.1, 600.0, .001)
s1 =  t1 * np.sin(np.pi / t1)
fig, ax2 = plt.subplots()
ax2.plot(t1, s1)
ax2.grid()
plt.show()
```

```python
# Data for plotting
t = np.arange(0.1, 16.0, .001)
s =  (4* t ** 2 - 1) / (2 * t ** 2 + 1)
fig, ax = plt.subplots()
ax.plot(t, s)
ax.grid()


t1 = np.arange(0.1, 600.0, .001)
s1 =  (4* t1 ** 2 - 1) / (2 * t1 ** 2 + 1)
fig, ax2 = plt.subplots()
ax2.plot(t1, s1)
ax2.grid()
plt.show()
```

```python
# Data for plotting
t = np.arange(1, 600, 1, dtype = np.uint64)


s = np.power(2, 1 + 1 / t) / np.power( 1 + 1 / t, t)
print(s)
fig, ax = plt.subplots()
ax.plot(t, s)
ax.grid()
```

```python

from scipy.special import factorial
# Data for plotting
t = np.arange(1, 6, 1)
s = np.power(2, t) / factorial(t, exact=False)
print(s)
fig, ax = plt.subplots()
ax.plot(t, s)
ax.grid()
```
